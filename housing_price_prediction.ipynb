```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data.info()
```

```python
data.dropna(inplace=True) # dropping all the rows with null values
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['median_house_value'], axis=1)
Y = data['median_house_value']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
train_data = X_train.join(Y_train)
```

```python
# Feature Engineering

train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)

# Using log to turn right skewed data into a bell curve
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity, dtype=int)).drop(['ocean_proximity'], axis = 1) # Feature Engineering

# Turning each categorical variable in  ocean_proximity in binary format by one hot encoding
```

```python
train_data['bedroom_ratio']= train_data['total_bedrooms'] / train_data['total_rooms'] # Feature Engineering
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households'] # Making new features using existing features
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(numeric_only=True), annot=True, cmap='YlGnBu')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Feature Engineering

scaler = StandardScaler()

# scaling ensures that numeric features are on a similar scale and prevents features with larger ranges from disproportionately affecting the model.

X_train, Y_train = train_data.drop(['median_house_value'], axis = 1), train_data['median_house_value']
x_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()

reg.fit(x_train_s, Y_train)
```

```python
test_data = X_test.join(Y_test)

# Feature Engineering

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)  # Using log to turn right skewed data into a bell curve

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity, dtype=int)).drop(['ocean_proximity'], axis=1) 

# Turning each categorical variable in  ocean_proximity in binary format by one hot encoding
                                                                                           
test_data['bedroom_ratio']= test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households'] # Making new features using existing features
```

```python
X_test, Y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
x_test_s = scaler.transform(X_test)
```

```python
reg.score(x_test_s, Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train_s, Y_train)
```

```python
forest.score(x_test_s, Y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()  

param_grid = {
    "n_estimators": [100, 200, 300],
    "min_samples_split": [2, 4],
    "max_depth": [None, 2, 4, 6]
}                                       # Hyperparameter tuning in random forest using k-fold cross validation

grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring = "neg_mean_squared_error",
                           return_train_score = True)

grid_search.fit(x_train_s, Y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(x_test_s, Y_test)
```
